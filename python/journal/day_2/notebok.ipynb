# Feature Selection

```python
import tensorflow_addons as tfa
```

```python
dataset = df.copy()
```

```python
def windowing(data, window_size, step_size):
    x = []
    y = []
    for i in range(0, len(data) - window_size, step_size):
        _x = np.array(data[i: i + window_size])
        _y = np.array([data.iloc[i + window_size]['target']])
        x.append(_x)
        y.append(_y)

    return np.array(x), np.array(y)


window_dataset = windowing(dataset, window_size=24, step_size=12)
```

```python
x_train, y_train = window_dataset
```

```python
print(f"Input Sample", x_train.shape)
print(f"Output Sample", y_train.shape)
```

```python
import tensorflow as tf

models = tf.keras.models
layers = tf.keras.layers

# one hot encoding
_y_train = tf.keras.utils.to_categorical(y_train - 1)

learning_rate = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.01,
    decay_steps=len(x_train) // 32,
    decay_rate=0.96,
    staircase=True)

early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=32, restore_best_weights=True)

base_line_model = models.Sequential()
base_line_model.add(layers.Bidirectional(layers.LSTM(64), input_shape=(24, x_train.shape[2])))
base_line_model.add(layers.Dense(64, activation='relu'))
base_line_model.add(layers.Dense(12, activation='softmax'))
base_line_model.compile(optimizer=tf.optimizers.Adam(learning_rate=learning_rate), loss='categorical_crossentropy', metrics=['accuracy', tfa.metrics.F1Score(num_classes=12)])
base_line_model.fit(x_train, _y_train, epochs=256, batch_size=32, callbacks=[early_stopping])
_y = base_line_model.predict(x_train)
```
